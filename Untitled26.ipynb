```python
from google.colab import files
import pandas as pd
import io

uploaded = files.upload()

uploaded_file_key = list(uploaded.keys())[0]

csv_data = uploaded[uploaded_file_key].decode('utf-8')

df = pd.read_csv(io.StringIO(csv_data))

print("CSV File Content:\n", df)
```

```python
print(df.columns)
```

**Summary Statistics**

```python
import numpy as np

# List of columns with numeric values
numeric_columns = ['operational_activities_kdollars', 'key_stakeholders_kdollars', 'training_kdollars',
                   'other_kdollars', 'internal_governance_kdollars', 'non_public_servants_kdollars',
                   'public_servants_kdollars', 'hospitality_kdollars', 'conference_fees_kdollars',
                   'minister_kdollars']

# Calculate summary statistics for each numeric column
summary_statistics = {}
for column in numeric_columns:
    # Filter out non-numeric values
    numeric_values = df[column].apply(pd.to_numeric, errors='coerce').dropna()
    if numeric_values.empty:
        print(f"No numeric data found in column '{column}'.")
        continue

    mean = np.mean(numeric_values)  # Calculate mean
    median = np.median(numeric_values)  # Calculate median
    std_dev = np.std(numeric_values)  # Calculate standard deviation
    min_val = np.min(numeric_values)  # Calculate minimum value
    max_val = np.max(numeric_values)  # Calculate maximum value
    quartiles = np.percentile(numeric_values, [25, 50, 75])  # Calculate quartiles

    # Store summary statistics for the column
    summary_statistics[column] = {
        'Mean': mean,
        'Median': median,
        'Standard Deviation': std_dev,
        'Minimum': min_val,
        'Maximum': max_val,
        'Quartiles': quartiles
    }

# Print summary statistics for each column
for column, stats in summary_statistics.items():
    print(f"{column}:")
    for stat, value in stats.items():
        if isinstance(value, np.ndarray):
            # Format quartiles as strings
            quartiles_str = ', '.join([f"${q:.2f}" for q in value])
            print(f"{stat}: {quartiles_str}")
        else:
            print(f"{stat}: ${value:.2f}")
    print()
```

```python
import matplotlib.pyplot as plt

# Plot histograms for selected numerical variables
plt.figure(figsize=(12, 8))
plt.hist(df['operational_activities_kdollars'], bins=20, color='skyblue', alpha=0.7, label='Operational Activities')
plt.hist(df['hospitality_kdollars'], bins=20, color='salmon', alpha=0.7, label='Hospitality')
plt.hist(df['conference_fees_kdollars'], bins=20, color='green', alpha=0.7, label='Conference Fees')
plt.title('Distribution of Expenditures')
plt.xlabel('Expenditure (Kdollars)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Box plots to visualize the distribution and identify outliers:**


```python
import seaborn as sns

# Plot box plots for selected numerical variables
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['operational_activities_kdollars', 'hospitality_kdollars', 'conference_fees_kdollars']])
plt.title('Box Plot of Expenditures')
plt.ylabel('Expenditure (Kdollars)')
plt.xlabel('Expenditure Type')
plt.show()
```

**Scatter plots to visualize relationships between numerical variables:**


```python
# Plot scatter plots for selected numerical variables
plt.figure(figsize=(12, 8))
sns.scatterplot(x='public_servants_kdollars', y='non_public_servants_kdollars', data=df)
plt.title('Scatter Plot of Public vs Non-Public Servants Expenditures')
plt.xlabel('Public Servants Expenditure (Kdollars)')
plt.ylabel('Non-Public Servants Expenditure (Kdollars)')
plt.show()
```

**Line plots to visualize trends over time**

```python
# Assuming 'year' is the column representing time
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='public_servants_kdollars', data=df)
sns.lineplot(x='year', y='non_public_servants_kdollars', data=df)
plt.title('Trends in Expenditures Over Time')
plt.xlabel('Year')
plt.ylabel('Expenditure (Kdollars)')
plt.legend(['Public Servants', 'Non-Public Servants'])
plt.show()
```
